# Similasyon ve Modelleme

```python
import matplotlib.pyplot as plt
import random as rnd
```

```python
Bilgisayar_Sayisi=[]

for i in range(0,18):  # Bu kod bloğunda 180 dk da 18 bilgisayar Tamire gemlmiştir anlamına geliyor.
    Bilgisayar_Sayisi.append(i)
print('Bilgisayar Sayıları: ',Bilgisayar_Sayisi)
```

```python
Kuyruk=[]
Dakikalar=0
Kuyruk_Sayisi=0
```

```python
a=[]
for i in range(3,21):              # 10 dk nın altında da tamir edilebilir 3 ie 21 ve 10 dk üzeri de tamir edilebilir.
    a.append(rnd.randrange(18))         # Ortalama 12dk olması 1dk arası tamir edilme aralığı mevcuttur.
print('rastgele tamir edilme zamanları: ',a) # Tamir süreleri değişir.
```

```python
for i in range(len(Bilgisayar_Sayisi)):
    if a[i]>10:
        Kuyruk.append(a[i]-10)  # Negatif değerler çıkamasın diye 10 ile karşılaştırdık.
        Dakikalar=Dakikalar+Kuyruk[i]
        Kuyruk_Sayisi=Kuyruk_Sayisi+1
    else:
        Kuyruk.append(0)
print("Toplam Dakika: ",Dakikalar)
```

```python
plt.figure(figsize=(16,6))
plt.title('GELEN BİLGİSAYARLARIN TAMİR EDİLME DURUMLARI VE ZAMANI')
plt.xlabel('Varış Zamanları')
plt.ylabel('Gelen bilgisayar Sayıları')
plt.bar(Bilgisayar_Sayisi,a, width=0.5, color="red", label="bilgisayar sayısı")
plt.show()
```

```python
plt.figure(figsize=(16,6))
plt.title('KUYRUKTA GEÇEN SÜRELER  ')
plt.xlabel('Varış Zamanları')
plt.ylabel('Gelen bilgisayar Sayıları')
plt.bar(Bilgisayar_Sayisi,Kuyruk, width=0.5, color="red", label="kuyrukta geçen süre")
plt.show()
```
