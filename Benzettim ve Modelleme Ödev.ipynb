# SİMÜLASYON VE MODELLEME 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

np.random.seed(0)
```

#  gelen bilgisayar ve işlem gören bilgisayar sayıları


```python
bilgisayar_sayisi=[]
k=1
        
for i in range(200): # Dakika olarak 180 yerine 200 vermemin nedeni 4 bilgisayar olan sistem zamanında hata vermesinden kaynaklandı
    if i%10==0 and i%12!=0:
        k=k+1  #10 dakikadan sonra Tamire  Gelen Bilgisayar Sayısını 1 arttır
        bilgisayar_sayisi.append(k)
    elif i%12==0 and i%10!=0:  
        k=k-1  #12 dk dan sonra bilgisayar sayısını 1 azalt, işlem gördüğü için 
        bilgisayar_sayisi.append(k)
        
    if i%10==0 and i%12==0:
        k=k
        bilgisayar_sayisi.append(k)

print("Her defasında gelen bilgisayar ve tamir edilen bilgisayar: ",bilgisayar_sayisi)

servis_ve_Varis_Zamani=[]
for i in range(200):
    if i%10==0 or i%12==0:
        servis_ve_Varis_Zamani.append(i)
print("Anlık Dakikalar :",servis_ve_Varis_Zamani)
        
```

```python

plt.figure(figsize=(16,6))
plt.title('GELEN BİLGİSAYARLARIN TAMİR EDİLME DURUMLARI VE ZAMANI')
plt.xlabel('Varış Zamanları')
plt.ylabel('Gelen bilgisayar Sayıları')
plt.plot(servis_ve_Varis_Zamani,bilgisayar_sayisi,marker='o')
plt.show()
```

# Sistemdeki ortalama Bilgisayar Sayıları

```python
T_1=0  # Bir bilgisayarın olduğu sistem zamanı
T_2=0  # iki bilgisayarın olduğu sistem zamanı
T_3=0  # üç bilgisayarın olduğu sistem zamanı
T_4=0  # dört bilgisayarın olduğu sistem zamanı

T_Bir=[] # Ortalama Bilgisayar Sayısı Dizisi.. Zaman Aralıkları
T_Iki=[]
T_Uc=[]
T_Dort=[]
```

```python


for i in range(0,31):
    if bilgisayar_sayisi[i]==1:
        
        T_Bir.append(servis_ve_Varis_Zamani[i+1]-servis_ve_Varis_Zamani[i])
        T_1+=servis_ve_Varis_Zamani[i+1]-servis_ve_Varis_Zamani[i]
          
             
    if bilgisayar_sayisi[i]==2:
        
        T_Iki.append(servis_ve_Varis_Zamani[i+1]-servis_ve_Varis_Zamani[i])
        T_2+=servis_ve_Varis_Zamani[i+1]-servis_ve_Varis_Zamani[i]
        
        
    if bilgisayar_sayisi[i]==3:
        
        T_Uc.append(servis_ve_Varis_Zamani[i+1]-servis_ve_Varis_Zamani[i])
        T_3+=servis_ve_Varis_Zamani[i+1]-servis_ve_Varis_Zamani[i]
        
    
    if bilgisayar_sayisi[i]==4:
        
        T_Dort.append(servis_ve_Varis_Zamani[i+1]-servis_ve_Varis_Zamani[i])
        T_4+=servis_ve_Varis_Zamani[i+1]-servis_ve_Varis_Zamani[i]
        
        
        
        
print("Bir bilgisayarın olduğu zaman aralıkları: ",T_Bir,"  Toplamı: ",T_1)
print('*************************************')
print("İki bilgisayarın olduğu zaman aralıkları: ",T_Iki,"  Toplamı: ",T_2)
print('*************************************')
print("Üç bilgisayarın olduğu zaman aralıkları: ",T_Uc,"  Toplamı: ",T_3)
print('*************************************')
print("Dört bilgisayarın olduğu zaman aralıkları: ",T_Dort,"  Toplamı: ",T_4)
```

# L DEĞERLERİ

```python
L=0
Toplam=0
Toplam=T_1*1+T_2*2+T_3*3+T_4*4
L=Toplam/180   # Üç saaat için ortalama hesapladık yani üç saate sistemde kaç tane bilgisayar var onu hesaplıyor

print(L," ortalama Bilgisayar sayısı")
```

# LQ(t) DEĞERLERİ

```python
TQ_1=0  # Bir bilgisayarın kuyrukta beklediği sistem zamanı..
TQ_2=0  # iki bilgisayarın  kuyrukta beklediği sistem zamanı..
TQ_3=0  # üç bilgisayarın  kuyrukta beklediği sistem zamanı..

TQ_Bir=[] # Ortalama kuyrukta bekleyen Bilgisayar Sayısı Dizisi.. Zaman Aralıkları
TQ_Iki=[]
TQ_Uc=[]
```

```python

for i in range(0,31):
    if bilgisayar_sayisi[i]==2:
        
        TQ_Bir.append(servis_ve_Varis_Zamani[i+1]-servis_ve_Varis_Zamani[i])
        TQ_1+=servis_ve_Varis_Zamani[i+1]-servis_ve_Varis_Zamani[i]
          
             
    if bilgisayar_sayisi[i]==3:
        
        TQ_Iki.append(servis_ve_Varis_Zamani[i+1]-servis_ve_Varis_Zamani[i])
        TQ_2+=servis_ve_Varis_Zamani[i+1]-servis_ve_Varis_Zamani[i]
        
        
    if bilgisayar_sayisi[i]==4:
        
        TQ_Uc.append(servis_ve_Varis_Zamani[i+1]-servis_ve_Varis_Zamani[i])
        TQ_3+=servis_ve_Varis_Zamani[i+1]-servis_ve_Varis_Zamani[i]
        

        
print("Bir bilgisayarın kuyrukta beklediği zaman aralıkları: ",TQ_Bir,"  Toplamı: ",TQ_1)
print('*************************************')
print("İki bilgisayarın kuyrukta beklediği zaman aralıkları: ",TQ_Iki,"  Toplamı: ",TQ_2)
print('*************************************')
print("Üç bilgisayarın kuyrukta beklediği zaman aralıkları: ",TQ_Uc,"  Toplamı: ",TQ_3)
print('*************************************')
```
